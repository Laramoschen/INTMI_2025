```python
#5 – Uma empresa deseja reajustar salários com base em faixas de rendimento. Receba o salário atual e aplique um aumento conforme as faixas (10%, 7% ou 5%), usando condicionais encadeadas. Exiba o novo salário.

salario_atual = float(input("Digite o salário atual: R$"))

if salario_atual > 2500:
  percentual_reajuste = 0.10
elif salario_atual > 1500:
  percentual_reajuste = 0.07
else:
  percentual_reajuste = 0.05

reajuste = salario_atual * percentual_reajuste
novo_salario = salario_atual + reajuste

print(f"O novo salário será de: R$ {novo_salario:.2f}")
```

```python
#11 – Uma calculadora avançada deve permitir que o usuário some quantos números quiser. Peça números repetidamente até que o valor 0 seja digitado. Ao final, mostre a soma total.

def calculadora_soma_avancada():

    soma_total = 0
    while True:
        try:

            numero_input = input("Digite um número para somar (ou 0 para terminar): ")
            numero = float(numero_input)


            if numero == 0:
                break

            soma_total += numero
            print(f"Soma parcial: {soma_total}")

        except ValueError:

            print("Entrada inválida. Por favor, digite um número válido.")


    print(f"\nSoma total final: {soma_total}")


calculadora_soma_avancada()
```

```python
#13 – Uma máquina de simulação deve somar números sequenciais começando em 1 até que a soma ultrapasse 100, mostrando cada número utilizado. Utilize um laço while.
soma = 0
numero = 1

while soma <= 100:
    print("Somando:", numero)
    soma += numero
    numero += 1

print("Soma final:", soma)
```

```python
#14 – Uma pesquisa está registrando idades. Continue lendo idades até que o usuário digite um valor negativo. Ao final, calcule e exiba a média das idades digitadas.
soma = 0
quantidade = 0

while True:
    idade = int(input("Digite uma idade (negativa para encerrar): "))

    if idade < 0:
        break

    soma += idade
    quantidade += 1

if quantidade > 0:
    media = soma / quantidade
    print("Média das idades:", media)
else:
    print("Nenhuma idade válida foi digitada.")
```

```python
#15 – Um cronômetro digital deve realizar uma contagem regressiva personalizada. Receba um número inicial e exiba a contagem até 0 usando um while.
inicio = int(input("Digite o número inicial da contagem: "))

while inicio >= 0:
    print(inicio)
    inicio -= 1

print("Contagem finalizada")
```

```python
#21 – Um app de produtividade quer identificar qual de quatro tarefas possui o nome mais longo. Leia quatro nomes de tarefas e exiba aquela com mais caracteres.
tarefas = []

for i in range(4):
    tarefa = input(f"Digite o nome da {i+1}ª tarefa: ")
    tarefas.append(tarefa)

# Inicializa a primeira tarefa como a maior
tarefa_mais_longa = tarefas[0]

for tarefa in tarefas[1:]:
    if len(tarefa) > len(tarefa_mais_longa):
        tarefa_mais_longa = tarefa

print("A tarefa com o nome mais longo é:", tarefa_mais_longa)
```

```python

]]#23 – Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.
positivos = 0
negativos = 0
zeros = 0

for i in range(10):
    numero = float(input(f"Digite o {i+1}º número: "))

    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print("Números positivos:", positivos)
print("Números negativos:", negativos)
print("Números iguais a zero:", zeros)
```

```python
#27 – Crie um sistema de login com limite de tentativas. O usuário terá no máximo 3 tentativas para digitar a senha correta. Caso acerte antes, use break para encerrar o laço.

senha_correta = "python123"
tentativas = 0
max_tentativas = 3

while tentativas < max_tentativas:
    senha = input("Digite a senha: ")

    if senha == senha_correta:
        print("Acesso permitido")
        break
    else:
        tentativas += 1
        print(f"Senha incorreta. Tentativa {tentativas} de {max_tentativas}.")

if tentativas == max_tentativas and senha != senha_correta:
    print("Número máximo de tentativas atingido. Acesso bloqueado.")
```

```python

#28 – Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.
soma_positivos = 0

for i in range(10):
    numero = float(input(f"Digite o {i+1}º número: "))

    if numero < 0:
        continue  # ignora números negativos

    soma_positivos += numero

print("Soma dos valores positivos:", soma_positivos)
```

```python
#29 – Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.
soma = 0
numero = 1

while True:
    soma += numero
    print(f"Adicionando {numero}, soma atual = {soma}")

    if soma > 200:
        print("A soma ultrapassou 200. Encerrando.")
        break

    numero += 1
```

```python
#30 – Um programa de edição quer contar quantas palavras o usuário digitou. Continue lendo palavras até que o usuário digite “fim”. Depois, exiba quantas palavras válidas foram informadas.

contagem = 0

while True:
    palavra = input("Digite uma palavra (ou 'fim' para encerrar): ")
    if palavra.lower() == "fim":
        break
    contagem += 1

print("Você digitou", contagem, "palavras válidas.")
```
