```python
#Peça um número de 1 a 7 e informe o dia da semana correspondente. Crie um algoritmo em Python que associe o número ao dia correto.
numero_dia = int(input("Digite um número de 1 a 7: "))
if numero_dia == 1:
    print("O dia correspondente é Domingo.")
elif numero_dia == 2:
    print("O dia correspondente é Segunda-feira.")
elif numero_dia == 3:
    print("O dia correspondente é Terça-feira.")
elif numero_dia == 4:
    print("O dia correspondente é Quarta-feira.")
elif numero_dia == 5:
    print("O dia correspondente é Quinta-feira.")
elif numero_dia == 6:
    print("O dia correspondente é Sexta-feira.")
elif numero_dia == 7:
    print("O dia correspondente é Sábado.")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python
#Peça um ano e informe se ele é bissexto. Crie um algoritmo em Python que verifique se o ano é divisível por 4 e não por 100, exceto se também for divisível por 400.
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"O ano {ano} é bissexto.")
else:
  print(f"O ano {ano} não é bissexto.")
```

```python
#Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.]
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    salario_base = 40 * valor_por_hora
    horas_extras = horas_trabalhadas - 40
    valor_horas_extras = horas_extras * valor_por_hora * 1.5
    salario = salario_base + valor_horas_extras

print(f"O salário a ser pago é: R${salario:.2f}")
```
